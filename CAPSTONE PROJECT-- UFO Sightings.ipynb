# UFO Sightings
## Objective- Explore UFO sighting reports from the National UFO Research Center (NUFORC).

### Import libraries:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Make a list of missing value types:

```python
missing_values = ["NaN", "--"]
```

### Import NUFORC dataset:

```python
import pandas as pd
df = pd.read_csv('https://query.data.world/s/ktbxzu4stmovbqwthrl52mhgt7xddi?dws=00000', na_values = missing_values)
```

### View data frame

```python
df
```

### View list of column labels:

```python
df.columns
```

### Use info() method to print data set details:

```python
df.info()
```

### Drop columns to reduce data set:

```python
drop_cols = ['summary','duration',
       'stats', 'report_link', 'text']
df.drop(drop_cols, inplace=True, axis=1)
```

### View first 5 rows of data to confirm columns were dropped:

```python
df.head()
```

### Rename column labels to improve readability:

```python
df = df.rename(columns = {'country':'Country', 'city':'City', 'state':'State', 'date_time':'Date_Time', 'shape':'Shape', 
       'posted':'Reported', 'city_latitude':'Latitude', 'city_longitude':'Longitude'})
df
```

### Add column for sighting 'Year':

```python
from datetime import datetime
df['Year'] = pd.DatetimeIndex(df['Date_Time']).year
df
```

### Add column for sighting 'Month':

```python
df['Month'] = pd.DatetimeIndex(df['Date_Time']).month
df
```

### Check for missing values:

```python
df.isnull().sum().sum()
```

### Drop missing values in place:

```python
df.dropna(inplace=True)
df
```

### Reset index of dataframe:

```python
df.reset_index(drop=True, inplace=True)
df
```

### Confirm missing values are dropped:

```python
df.isnull().sum().sum()
```

### Check dtypes:

```python
df.info()
```

### Convert Year and Month columns to integer:

```python
df= df.astype({"Year":'int', "Month":'int'})
```

### Confirm Year and Month columns converted to integer (decimal-zero is now removed):

```python
df
```

### Get sum of UFO sightings by Country and Year

```python
df.groupby(['Country','Year'])['Reported'].count()
```

### Noticed location names were not uniform (some are capitalized, some are not). Convert Country, State, City columns from Object to String:

```python
df['Country'].astype(str).inplace=True
df['City'].astype(str).inplace=True
df['State'].astype(str).inplace=True
```

### Convert first character of each word to uppercase and remaining to lowercase (Country, State, and City):

```python
df['Country'].str.title().inplace=True
df['City'].str.title().inplace=True
df['State'].str.title().inplace=True
df
```

### Filter out variables that do not apply to 'Shape':

```python
df = df.loc[((df['Shape'] == 'changing') | (df['Shape'] == 'chevron') | (df['Shape'] == 'cigar') | (df['Shape'] == 'circle') | (df['Shape'] == 'cone') | (df['Shape'] == 'cross') | (df['Shape'] == 'cylinder') | (df['Shape'] == 'delta') | (df['Shape'] == 'diamond') | (df['Shape'] == 'disk')
             | (df['Shape'] == 'egg') | (df['Shape'] == 'fireball') | (df['Shape'] == 'flash') | (df['Shape'] == 'formation')
             | (df['Shape'] == 'light') | (df['Shape'] == 'other') | (df['Shape'] == 'oval') | (df['Shape'] == 'rectangle')
             | (df['Shape'] == 'sphere') | (df['Shape'] == 'star') | (df['Shape'] == 'teardrop') | (df['Shape'] == 'triangle') | (df['Shape'] == 'unknown'))]
df
```

### Count UFO sightings by Year and sort descending:

```python
df.groupby(['Year'])['Reported'].count().sort_values(ascending=False)
```

### Total UFO sightings by ascending Year:

```python
df.groupby(['Year'])['Reported'].count()
```

### Visualize UFO sightings by Year using line plot:

```python
df.groupby(['Year'])['Reported'].count().plot(kind='line',x='Year',y='Reported')
plt.title("UFO Sightings by Year")
plt.ylabel("Reports")
plt.xlabel("Year")
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

### Visualize UFO sightings by Year using bar plot:

```python
year= df.groupby(['Year'])['Reported'].count().reset_index(name='counts')
sns.barplot (x='Year',y='counts',data=year)
plt.gca().set_xticklabels(year['Year'], rotation=60, horizontalalignment= 'right')
plt.title("UFO Sightings by Year", fontsize = 20)
plt.ylabel("Reports", fontsize = 15)
plt.xlabel("Year", fontsize = 15)
plt.show()
```
